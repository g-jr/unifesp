```python
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import random
N = 100 # quantidade de vertices
grau = 4 # grau medio do grafo
```

## Introdução
Através da implementação de um grafo de Redes Aleatórias (Erdös-Rényi), é buscado a implementação de um modelo de propagação de doença em rede através de uma probabilidade de contaminação entre os indivíduos.<br>
Esse modelo de Rede Aleatórias é dado pela função _G(n, m)_, n é a quantidade de vértices e m é a quantidade de arestas inseridas de maneira aleatória no grafo.

## Geração da rede
É utilizado um grafo com N vértices e E arestas. A quantidade de arestas é calculada com base na equação de grau medio igual à 4:
$$grau = \frac{2\cdot E}{N}$$
<br>
Logo, temos uma quantidade de 100 vértices e 200 arestas e um grau médio 4.<br>
Esse grafo – criado a partida da biblioteca networkx – é convertido em uma matriz de adjacência para à implementação da propagação.

```python
E = (grau * N) / 2
G = nx.gnm_random_graph(N, E).to_undirected()
nx.draw_random(G, with_labels=True)
```

```python
matriz = np.asarray(nx.to_numpy_matrix(G, dtype=int))
plt.matshow(matriz)
plt.show()
```

## Dinâmica (Propagação)
Cada vértice representa um indivíduo no grafo, e as arestas representam seus vizinhos.<br>
É definida uma constante de probabilidade de contaminação, comum a todo os vértices, que sera utilizada quando o estado de algum estiver como contaminado (estado = 1). Essa contaminação tem uma duração definida pela constante D – conforme cada iteração a duração é diminuida. Um vértice contaminado podera ser contaminado novamente após a expiração desse prazo.<br>
### Parâmetros
__T__: quantidade de iterações (tempo)<br>
__D__: quantidade de duração<br>
__prob__: probabilidade de contaminação<br>


```python
class RedePropagacao(object):
  def __init__(self, matriz):
    self.matriz = matriz
    self.T = 100
    
  def setParametros(self, duracao, quantidade, prob): # seta os parametros da rede
    self.D = duracao
    self.quantidade = quantidade
    self.prob = prob
    
  def contagiar(self): # contagia os elementos da rede
    arr_contaminados = [random.randint(0, N) for _ in range(self.quantidade)]
    for i in arr_contaminados:
      self.estado[i] = 1
      self.duracao[i] = self.D
      
  def plotar(self, t):
    fig, axis = plt.subplots()
    axis.imshow(np.vstack((self.estado, self.estado)), aspect='auto', vmin=0, vmax=1, cmap='coolwarm')
    axis.set_axis_off()
    fig.subplots_adjust(top=1, bottom=0.01, left=0, right=5)
    fig.text(0,0.5, "T: %04d"%(t), va='center', ha='right', fontsize=50)
    plt.show()
    
  def propagar(self):
    self.estado = np.zeros((N), dtype=int)
    self.duracao = np.zeros((N), dtype=int)
    self.contagiar()
    self.plotar(0)
    
    for t in range(1, self.T):
      new_estado = np.zeros((N), dtype=int)
      for n in range(N):
        if self.duracao[n] > 0:
          self.duracao[n] -= 1 # time-step
        indices = np.where(self.matriz[n,:]==1)[0] # lista de adjacência nó n
        viz_contaminado = False
        for idx in indices:
          if self.estado[idx] == 1: # vizinho está doente
            viz_contaminado = True
            break
        if np.random.random_sample() <= self.prob and viz_contaminado: # chance de contágio    
          new_estado[n] = 1
          self.duracao[n] = D
        else:
          if self.duracao[n] == 0:
            new_estado[n] = 0
          else:
            new_estado[n] = self.estado[n]

      self.estado = new_estado

      self.plotar(t)
      if len(np.nonzero(self.estado)[0]) == 0:
        break
```

```python
rede = RedePropagacao(matriz)
```
