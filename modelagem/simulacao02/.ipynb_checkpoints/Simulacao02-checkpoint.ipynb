## Introdução

-Conceito de rede trófica<br>
-Modelo de Lotka-Volterra

### Equações
Grama(Produtor):<br>
dG = G(t) * (A1 * (1 - G(t)/K) - A2 * Z(t) - A3 * I(t)) * dT<br>
<br>
Zebra - herbívoro (consumidor primário):<br>
dZ = Z(t) * (B1 * G(t) - B2 *  L(t) - B3 * H(t) - B4 * Z(t))* dT<br>
<br>
Impala - herbívoro (consumidor primário):<br>
dI = I(t) * (C1 * G(t) - C2 * L(t) - C3 * H(t) - C4 * I(t)) * dT<br>
<br>
Leão - carnívoro (consumidor secundário):<br>
dL = L(t) * (D1 * Z(t) + D2 * I(t) - D3 * L(t)) * dT<br>
<br>
Hiena - carnívoro (consumidor secundário):<br>
dH = H(t) * (E1 * Z(t) + E2 * I(t) - E3 * H(t)) * dT<br>
<br>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Parâmetros
G0: População inicial de Grama.<br>
Z0: População inicial de Zebra.<br>
I0: População inicial de Impala.<br>
L0: População inicial de Leão.<br>
H0: População inicial de Hiena.<br>

```python
DT = 0.01
TMax = 200

G0 = 400
Z0 = 100
I0 = 120
L0 = 60
H0 = 40

G = G0
Z = Z0
I = I0
L = L0
H = H0

####---constantes---####
K = 3600 # capacidade do meio

# Grama
A1 = 73.34 # taxa de crescimento
A2 = 0.03 # consumo por zebras
A3 = 0.035 # consumo por impalas

# Zebra
B1 = 0.034 # consumo de grama
B2 = 0.005 # morte por leoes
B3 = 0.009 # morte por hienas
B4 = 0.003# constante de mortalidade

# Impala
C1 = 0.034 # consumo de grama
C2 = 0.009 # morte por leoes
C3 = 0.005 # morte por hienas
C4 = 0.001 # constante de mortalidade

# Leao
D1 = 0.012 # consumo de zebras
D2 = 0.015 # consumo de impalas
D3 = 0.015 # constante de mortalidadere

# Hiena
E1 = 0.017 # consumo de zebras
E2 = 0.013 # consumo de impalas
E3 = 0.015 # constante de mortalidade
####---fim constantes---####
t_array = [0]
g_array = [G]
z_array = [Z]
i_array = [I]
h_array = [H]
l_array = [L]
for t in np.arange(DT, TMax, DT):
  Dg = G * ((A1*(1 - G/K)) - A2*Z - A3*I)*DT
  Dz = Z * (B1*G - B2*L - B3*H - B4*Z)*DT
  Di = I * (C1*G - C2*L - C3*H - C4*I)*DT
  Dl = L * (D1*Z + D2*I - D3*L)*DT
  Dh = H * (E1*Z + E2*I - E3*H)*DT
  G += Dg
  Z += Dz
  I += Di
  L += Dl
  H += Dh
  t_array.append(t)
  g_array.append(G)
  z_array.append(Z)
  i_array.append(I)
  h_array.append(H)
  l_array.append(L)
dados = pd.DataFrame({'tempo': t_array, 'grama': g_array, 'zebra': z_array, 'impala': i_array, 'hiena': h_array, 'leao': l_array})
dados.set_index('tempo', inplace=True)
```

```python
plt.plot(dados)
plt.xlabel('tempo', fontsize=15)
plt.ylabel('animais', fontsize=15)
plt.legend(dados.columns.values, loc='upper right')
plt.rcParams["figure.figsize"] = [25,9]
plt.xlim([75,80])
plt.show()
```
