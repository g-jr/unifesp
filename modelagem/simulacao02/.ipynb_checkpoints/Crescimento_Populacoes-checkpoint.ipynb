## 1. Modelo Malthusiano

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DT = 0.01
TMax = 200
N0 = 1
alpha = 0.2

n = N0
n_array = [N0]
t_array = [0]

for t in np.arange(DT, TMax, DT):
    Dn = alpha*n*DT
    n = n + Dn
    n_array.append(n)
    t_array.append(t)
```

```python
plt.plot(t_array, n_array)
plt.title('População - Crescimento Exponencial')
plt.show()
```

## 2. Modelo Logístico

```python
DT = 0.01
TMax = 100
N0 = 1
K = 150
alpha = 0.2

n = N0
n_array = [N0]
t_array = [0]

for t in np.arange(DT, TMax, DT):
    Dn = alpha*n*(1.0 - n/K)*DT
    n = n + Dn
        
        
    if t==50:
        n = n*2
        
    n_array.append(n)
    t_array.append(t)
```

```python
plt.plot(t_array, n_array)
plt.title('População - Crescimento Limitado')
plt.show()
```
